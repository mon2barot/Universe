```python
data = df[['contact','day','month','duration','campaign','previous','pdays','customer']]
data.sample(10)
```

Replace categorical values with 0 and 1

```python
d = {'no':0, 'yes':1}
data.customer = data.customer.map(d)
```

Goal: who will be a customer in the future, therefore target is customer and predictors are all other remaining variables

```python
x = data.drop(['customer'],axis =1)
y = df[['customer']]
```

```python
y.sample(5)
```

```python
x.sample(5)
```

creating dummy variables for contact and month

```python
x_dummies = pd.get_dummies(x, drop_first=True)

x_dummies.head()
```

```python
x_dummies.info()
```

Splitting the data into training and testing subsets

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_dummies, y, test_size = .20, random_state=0)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

Applying RandomForest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r = RandomForestClassifier(random_state=0)
r.fit(xtrain,ytrain)
```

Predicting future customers

```python
r_pred = r.predict(xtest)
r_pred
```

Evaluating the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(r_pred,ytest)
```
